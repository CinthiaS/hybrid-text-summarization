```
import re

'''
    Input:  Texto
    Output: Texto sem stopwords
'''
def remove_stopwords(line):
  stop_words = set(stopwords.words('english'))
  tokens = word_tokenize(line)
  texto = [i for i in tokens if not i in stop_words]
  return texto

def preprocess(file):
    processed = [] # preprocess the file
    lines = file.read().split('\n')
    for line in lines:
        line = line.lower() # lower case
        line = line.replace('#','') # replace the token '#' with '<num>'
        line = line.replace(',','')
        line = line.replace('. sub .','')
        line = line.replace('. sup .','')
        line = line.replace('and and','and')
        line = line.replace('-','')
        #Remove strings compostas por caracteres unicos
        line= re.sub(r'\b\w\b', ' ', line)
        line = re.sub(r'\d+', '', line)
        text = re.sub(r"[()`#/@';:%<>$&\"{}~+=?|]", " ", line) # replace other tokens with a space
        text = text.rstrip().strip() # strip white space
        text = remove_stopwords(line)
        text = ' '.join(text)
        processed.append(text)

    return processed

train_article_path = "/content/drive/My Drive/Colab Notebooks/sumdata/database_td/np/resumo.train.txt"
#train_title_path = "/content/drive/My Drive/Colab Notebooks/sumdata/database_td/np/titulo.train.txt"
valid_article_path = "/content/drive/My Drive/Colab Notebooks/sumdata/database_td/np/resumo.valid.txt"
#valid_title_path = "/content/drive/My Drive/Colab Notebooks/sumdata/database_td/np/titulo.valid.txt"

article_train = open(train_article_path,'r') 
#title_train = open(train_title_path,'r') 
article_valid = open(valid_article_path,'r') 
#title_valid = open(valid_title_path,'r') 

# preprocess the text
lines_article_train = preprocess(article_train)
#lines_title_train = preprocess(title_train)
lines_article_valid = preprocess(article_valid)
#lines_title_valid = preprocess(title_valid)

arq = open("/content/drive/My Drive/Colab Notebooks/sumdata/database_td/sem_stopwords/resumo.train.pp.txt","w")
for i in lines_article_train:
  arq.write(i + "\n")
arq.close()

'''
arq = open("/content/drive/My Drive/Colab Notebooks/sumdata/database_td/sem_stopwords/titulo.train.pp.txt","w")
for i in lines_title_train:
  arq.write(i + "\n")
arq.close()
'''
arq = open("/content/drive/My Drive/Colab Notebooks/sumdata/database_td/sem_stopwords/resumo.valid.pp.txt","w")
for i in lines_article_valid:
  arq.write(i + "\n")
arq.close()
'''
arq = open("/content/drive/My Drive/Colab Notebooks/sumdata/database_td/sem_stopwords/titulo.valid.pp.txt","w")
for i in lines_title_valid:
  arq.write(i + "\n")
arq.close()
'''
```
